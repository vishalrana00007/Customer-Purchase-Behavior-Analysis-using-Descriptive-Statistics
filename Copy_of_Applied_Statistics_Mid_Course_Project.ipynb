```python

from scipy import stats

# Calculating basic statistics for numerical variables
numerical_vars = [ 'MntFishProducts',
                  'MntMeatProducts', 'MntFruits', 'MntSweetProducts',
                  'MntWines', 'MntGoldProds', 'NumDealsPurchases',
                  'NumCatalogPurchases', 'NumStorePurchases',
                  'NumWebPurchases', 'NumWebVisitsMonth', 'Recency']

for var in numerical_vars:
  print(f"Variable: {var}")
  # Descriptive statistics
  print(sm_df[var].describe())

  # Checking for normality
  try:
    # Shapiro-Wilk test for normality
    stat, pval = stats.shapiro(sm_df[var])
    if pval > 0.05:
      print(f"{var} distribution might be normal (p-value: {pval:.2f})")
    else:
      print(f"{var} distribution might not be normal (p-value: {pval:.2f})")
  except:
    pass  # Handling potential errors for non-numeric data

# Analyzing Response (target) - Binomial distribution (potentially)
response_rate = sm_df['Response'].mean()
print(f"\nResponse (target): Binomial distribution (success rate: {response_rate:.2f})")

# Analyzing Complain - Binomial distribution (potentially)
complaint_rate = sm_df['Complain'].mean()
print(f"\nComplain: Binomial distribution (complaint rate in last 2 years: {complaint_rate:.2f})")
```

This summary provides a comprehensive overview of customer spending habits across various categories. ```python

```python
import scipy.stats as stats

# Selecting variables to analyze
selected_variables = ['NumStorePurchases', 'NumWebPurchases', 'Income']

# Checking normality assumption for selected variables using Shapiro-Wilk test
for var in selected_variables:
    stat, p = stats.shapiro(sm_df[var])
    print(f'{var} Shapiro-Wilk test p-value: {p}')

# Identifying variables that could follow specific distributions
# Example: If p-value is not significant for Normality assumption, 'NumStorePurchases' might follow a Poisson distribution

# Calculating probabilities and expected values based on the identified distribution (e.g., Poisson)
lambda_val = sm_df['NumStorePurchases'].mean()  # Expected value for Poisson distribution

# Calculating probabilities for Poisson distribution
poisson_probabilities = [stats.poisson.pmf(k, lambda_val) for k in range(0, 10)]  # Probabilities for 0 to 9 purchases

# Display probabilities for Poisson distribution
for k, prob in enumerate(poisson_probabilities):
    print(f'P(X = {k}) = {prob}')
```

# **Binary Variables (Bernoulli Probability Distribution):**

```python
# Calculating the proportion (probability) of customers who accepted the offer (1s)
accepted_offers_proportion = sm_df["Response"].mean()

# Printing the estimated proportion
print(f"Estimated Proportion of Customers who Accepted the Offer (p): {accepted_offers_proportion}")

# Plotting the probability mass function (PMF) for the Bernoulli distribution
plt.bar([0, 1], [1-accepted_offers_proportion, accepted_offers_proportion], tick_label=["Rejected (0)", "Accepted (1)"])
plt.title('Bernoulli Distribution of Offer Acceptance')
plt.xlabel('Offer Acceptance')
plt.ylabel('Probability')
plt.show()
```

```python

# Calculating the proportion (probability) of customers who complained (1s)
complaints_proportion = sm_df["Complain"].mean()

# Plotting the probability mass function (PMF) for the Bernoulli distribution
plt.bar([0, 1], [1-complaints_proportion, complaints_proportion], tick_label=["No Complaint (0)", "Complaint (1)"])
plt.title('Bernoulli Distribution of Customer Complaints')
plt.xlabel('Complaint Status')
plt.ylabel('Probability')
plt.show()
```

**Conclusion:** Analyzing offer acceptance through a Bernoulli distribution reveals a high probability of rejection. Conversely, the Bernoulli distribution of customer complaints suggests a very low likelihood of complaints arising. These insights are valuable for the company to prioritize areas for improvement and ultimately boost sales.

# **Continuous Variables (Normal Probability Distribution):**

```python
from scipy.stats import norm

# Calculating mean and standard deviation for Recency
mean_recency = sm_df["Recency"].mean()
std_dev_recency = sm_df["Recency"].std()

     # Printing the estimated mean and standard deviation
print(f"Estimated Mean Recency: {mean_recency}")
print(f"Estimated Standard Deviation of Recency: {std_dev_recency}")

     # Generating some data points for the x-axis
x_recency = np.linspace(mean_recency - 3*std_dev_recency, mean_recency + 3*std_dev_recency, 100)

     # Calculating the corresponding y-axis values using the normal distribution
y_recency = norm.pdf(x_recency, mean_recency, std_dev_recency)

# Plotting the probability distribution curves
plt.plot(x_recency, y_recency, label='Normal Distribution')
plt.title('Probability Distribution of Recency')
plt.xlabel('Days Since Last Purchase')
plt.ylabel('Probability density')
plt.legend()
plt.show()
```

**Conclusion:** Examining the normal probability distribution of recency reveals a key insight: purchase frequency peaks at around 50 days. This valuable customer behavior data presents the company with a strategic opportunity to optimize their sales strategies. By understanding this timeframe, the company can develop targeted campaigns or incentives to encourage more frequent purchases and potentially shorten the repurchase cycle.

# **K-Means Clustering**

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans


# Selecting relevant features (numerical for KMeans initially)
features = ['Year_Birth', 'MntFishProducts', 'MntMeatProducts', 'MntFruits',
            'MntSweetProducts', 'MntWines', 'MntGoldProds', 'NumDealsPurchases',
            'NumCatalogPurchases', 'NumStorePurchases', 'NumWebPurchases']

# Standardizing features (optional but recommended for KMeans)
scaler = StandardScaler()
scaled_features = scaler.fit_transform(sm_df[features])

# Elbow method to determine optimal number of clusters (potentially)
inertias = []
for k in range(1, 11):
  kmeans = KMeans(n_clusters=k, random_state=42)
  kmeans.fit(scaled_features)
  inertias.append(kmeans.inertia_)

# Plotting elbow method to visualize (optional)
import matplotlib.pyplot as plt
plt.plot(range(1, 11), inertias, '-o')
plt.xlabel("Number of clusters")
plt.ylabel("Inertia")
plt.show()

```

As it is not very clear from the elbow method that which value of K to choose.We have to find silhouette score to determine optimal numbers of clusters.

**Silhouette score**

```python

#Data imputation
imputer = SimpleImputer(strategy='mean')
sm_df_imputed = imputer.fit_transform(sm_df)

# Calculate silhouette scores for each cluster
silhouette_scores = []
for i in range(2, 10):
    kmeans = KMeans(n_clusters=i, random_state=42)
    cluster_labels = kmeans.fit_predict(sm_df_imputed)
    silhouette_scores.append(silhouette_score(sm_df_imputed, cluster_labels))

# Plot the silhouette scores
plt.bar(range(2, 10), silhouette_scores)
plt.xlabel('Number of clusters', fontsize=20)
plt.ylabel('Silhouette Score', fontsize=20)
plt.show()
```

```python
sc=max(silhouette_scores)
number_of_clusters=silhouette_scores.index(sc)+2
print("Number of Cluster Required is : ", number_of_clusters)
```
